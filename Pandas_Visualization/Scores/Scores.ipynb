```python
work_data = pd.DataFrame({"first_name":["Jason","Molly","Tine","Jake","Amy"],"last_name":["Miller","Jacobson","Ali","Milner","Cooze"],
                         "age":[42,52,36,24,73],"female":[0,1,1,0,1],"preTestScore":[4,24,31,2,3],"postTestScore":[25,94,57,62,70]})
```
